```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

# Read the CSV file into a DataFrame
df = pd.read_csv("dataset.csv")

def get_yz(df, x_min, x_max):
    filtered_df = df[(df['class'] == 'b') & (df['x'] >= x_min) & (df['x'] <= x_max)]
    y = filtered_df['y'].values
    z = filtered_df['z'].values
    return y, z

def plot_yz(x_min):
    x_max = x_min + 0.05
    y, z = get_yz(df, x_min, x_max)
    plt.scatter(y, z)
    plt.title('x = ' + str(x_min))
    plt.savefig('figs/scatter' + str(x_min*100) + '.png')
    plt.show()

for i in range(0, 20):
    j = i / 20
    plot_yz(j)

#interact(plot_yz, x_min=(df['x'].min(), df['x'].max(), 0.05))
```
